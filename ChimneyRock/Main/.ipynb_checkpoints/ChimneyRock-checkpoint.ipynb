```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix

%matplotlib inline
```

```python
img_lao = []
img_rao = []
num_lao = 8
num_rao = 8

for i in range(num_lao):
    filename = 'CAG/p000'+str(i)+'l.bmp'
    img = Image.open(filename)
    img_lao.append(img)
    
for i in range(num_rao):
    filename = 'CAG/p000'+str(i)+'r.bmp'
    img = Image.open(filename)
    img_rao.append(img)
```

```python
img_rao[1]
```

```python
def extractData(image):
    x = image.size[0]
    y = image.size[1]
    
    matrix = []

    for i in range(x):
        matrix.append([])
        for j in range(y):
            r, g, b = image.getpixel((j, i))            
            value = r * 299.0/1000 + g * 587.0/1000 + b * 114.0/1000
            value = int(value)
            matrix[i].append(value)
    return matrix
```

```python
data_rao = []
data_lao = []

for case_rao in img_rao:    
    data_case = extractData(case_rao)
    case = []
    for row in data_case:
        for column in row:
            case.append(column)            
    data_rao.append(case)
    
for case_lao in img_lao:    
    data_case = extractData(case_lao)
    case = []
    for row in data_case:
        for column in row:
            case.append(column)
    data_lao.append(case)
```

## Support Vector Classification

```python
num_train = 7
num_test = 2
c = [1] * num_train + [2] * num_train
```

```python
c
```

### Training

```python
X = np.array(data_lao[:num_train] + data_rao[:num_train])
y = np.array(c)
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

### Prediction

```python
y_predict = clf.predict(data_lao[-num_test:] + data_rao[-num_test:])
```

```python
img_test_l = Image.open('CAG/p0008l.bmp')
img_test_r = Image.open('CAG/p0008r.bmp')
```

```python
data_test_l = extractData(img_test_l)
case_test_l = []
for row in data_test_l:
    for column in row:
        case_test_l.append(column)
            
data_test_r = extractData(img_test_r)
case_test_r = []
for row in data_test_r:
    for column in row:
        case_test_r.append(column)
```
