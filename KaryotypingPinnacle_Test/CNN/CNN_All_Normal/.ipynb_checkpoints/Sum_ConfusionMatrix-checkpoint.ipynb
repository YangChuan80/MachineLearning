# Load Confusion Matrix

```python
confusion_all_list = []

for iterationNumber in range(7):
    confusion_all_list.append(pd.read_csv('ConfusionMatrix_KaryoType_CNN' + str(iterationNumber) + '.csv', index_col=0))
```

```python
confusion_sum = np.array(confusion_all_list[0])
```

```python
for i in range(6):
    confusion_sum = confusion_sum + confusion_all_list[i+1]
```

# Plot Heatmap

```python
plt.figure(figsize=(30, 24))
heat_map = sb.heatmap(confusion_sum, annot=True,  # It's a Seaborn function
                      cmap='coolwarm', 
                      xticklabels=confusion_sum.columns, yticklabels=confusion_sum.index, 
                      linewidths=.5, 
                      annot_kws={"size": 20})
sb.set(font_scale = 2)
```
