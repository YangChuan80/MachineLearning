```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
%matplotlib inline
```

```python
img = Image.open('CAG/p0000l.bmp')
```

```python
img
```

```python
lao = []
rao = []
for i in range(8):
    filename = 'CAG/p000'+str(i)+'l.bmp'
    img = Image.open(filename)
    lao.append(img)
    
for i in range(8):
    filename = 'CAG/p000'+str(i)+'r.bmp'
    img = Image.open(filename)
    rao.append(img)    
```

```python
rao[1]
```

```python
def extractData(image):
    x = image.size[0]
    y = image.size[1]
    
    matrix = []
    points = []
    integrated_density = 0

    for i in range(x):
        matrix.append([])
        for j in range(y):
            r, g, b = image.getpixel((j, i))            
            value = r * 299.0/1000 + g * 587.0/1000 + b * 114.0/1000
            value = int(value)
            matrix[i].append(value)
    return matrix
```

```python
data_rao = []
data_lao = []

for rao_case in rao:    
    data_case = extractData(rao_case)
    case = []
    for row in data_case:
        for column in row:
            case.append(column)            
    data_rao.append(case)
    
for lao_case in lao:    
    data_case = extractData(lao_case)
    case = []
    for row in data_case:
        for column in row:
            case.append(column)
    data_lao.append(case)
```

## Support Vector Classification

```python
c = [1]*8 + [2]*8
```

```python
c
```

```python
X = np.array(data_lao + data_rao)
y = np.array(c)
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

```python
print(clf.predict([data_lao[3]]))
```

```python
img_test_l = Image.open('CAG/p0008l.bmp')
img_test_r = Image.open('CAG/p0008r.bmp')
```

```python
data_test_l = extractData(img_test_l)
case_test_l = []
for row in data_test_l:
    for column in row:
        case_test_l.append(column)
            
data_test_r = extractData(img_test_r)
case_test_r = []
for row in data_test_r:
    for column in row:
        case_test_r.append(column)
```

```python
print('w = ',clf.coef_)
print('b = ',clf.intercept_)
print('Indices of support vectors = ', clf.support_)
print('Support vectors = ', clf.support_vectors_)
print('Number of support vectors for each class = ', clf.n_support_)
print('Coefficients of the support vector in the decision function = ', np.abs(clf.dual_coef_))
```
